###Example 4

```
#Note y=+/-1
def make_rect(X, x_left, x_right, y_bottom, y_top):
  n = X.shape[0]
  y = np.ones(n)
  for i in range(n):
    if x_left<X[i][0]<x_right and y_bottom<X[i][1]<y_top:
      y[i]=-1
  return y
```

```
#Example 4a
np.random.seed(17)
X = np.random.randn(100,2)
y = make_rect(X, -1, 1,-0.5, 0.5 )

plt.scatter(X[y==1][:,0], X[y==1][:,1])
plt.scatter(X[y==-1][:,0], X[y==-1][:,1])


plt.grid(True)
plt.show()
```

```
temp = RTTree()
temp.fit(X, y.astype(int))
# X: np.array (n, 2), y: np.array (n,) (y consists from -1 and 1)
if not temp.tree_struct['terminal']:
  plot_2D(X, temp.tree_struct['Wopt'], y, temp, title='3d')
else:
  print('No separating lines!')
```

```
#Example 4b
np.random.seed(17)
X = np.random.randn(100,2)
y = make_rect(X, -1, 1, -1, 1)

plt.scatter(X[y==1][:,0], X[y==1][:,1])
plt.scatter(X[y==-1][:,0], X[y==-1][:,1])

plt.grid(True)
plt.show()
```

```
temp = RTTree()
temp.fit(X, y.astype(int))
# X: np.array (n, 2), y: np.array (n,) (y consists from -1 and 1)
if not temp.tree_struct['terminal']:
  plot_2D(X, temp.tree_struct['Wopt'], y, temp, title='3d')
else:
  print('No separating lines!')
```

###Example 5

```
#Note y=+/-1
def make_concentric(X, r):
  n = X.shape[0]
  y = np.ones(n)
  for i in range(n):
    if np.linalg.norm(X[i]) > r:
      y[i]=-1
  return y
```

```
#Example 5a
np.random.seed(17)
X = np.random.randn(100,2)
y = make_concentric(X, 0.8)

plt.scatter(X[y==1][:,0], X[y==1][:,1])
plt.scatter(X[y==-1][:,0], X[y==-1][:,1])

plt.grid(True)
plt.show()
```

```
temp = RTTree()
temp.fit(X, y.astype(int))
# X: np.array (n, 2), y: np.array (n,) (y consists from -1 and 1)
if not temp.tree_struct['terminal']:
  plot_2D(X, temp.tree_struct['Wopt'], y, temp, title='3d')
else:
  print('No separating lines!')
```
